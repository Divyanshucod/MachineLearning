#### Data Manipulation with Pandas

```python
import pandas as pd

df = pd.read_csv('data.csv')
# print(df)

df.head(5)
```

```python
## Handling missing values
df.isnull()
## any it will give you true and false on the basis of column names 
df.isnull().any()
## sum of all missing values
df.isnull().sum()
## fill missing values
filled_values = df.fillna(0)
## when get missing values then filled that with mean of that column
df['sales_fiilValue'] = df['Sales'].fillna(df['Sales'].mean())
print(df)
```

```python
## Renaming column
df = df.rename(columns={'Sales':'Monthly Sales'})
# print(df)

df.dtypes
# changing datatype -> float64 to int64
df['Profit_Value'] = df['Profit'].astype(int)
# df.head()

## operations like increasing my multiplying or adding somevalue
df['Doubled_Value'] = df['Monthly Sales'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregation and grouping

grouped_mean = df.groupby('Product ID')['Monthly Sales'].mean()
print(grouped_mean)
```

```python
## apply multiple aggrigator functions
grouped_agg = df.groupby('Product ID')['Monthly Sales'].agg(['mean','sum','median'])
print(grouped_agg)
```

```python
## Merging and Joining DataFrame
df1 = pd.DataFrame({'Key':['A','B','C'],'Values':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','F'],'Values':[4,5,6]})

## merge dataframe based on 'Key column' 
pd.merge(df1,df2,on='Key',how='inner')
## with outer join
pd.merge(df1,df2,on='Key',how='outer')
```
