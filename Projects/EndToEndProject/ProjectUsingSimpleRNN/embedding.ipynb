#### Creating wordEmbedding Layer

```python
### Via OHE
from tensorflow.keras.preprocessing.text import one_hot
```

```python
Sentences = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tree',
    'I am a good boy',
    'I am a good developer',
    'Userstand the meaning of words',
    'i will become a business Man'
]
```

```python
## vocabulary size
voc_size = 10000
```

```python
one_hot_rep = [one_hot(words,voc_size) for words in Sentences]
```

```python
one_hot_rep
```

```python
##word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## so right know we have a sentences with different lengths so first of all make them as same size and to do that we can take maximun length(specifies the vocabulary) and we can add zeros
## either at the front using 'pre' or at end using 'post' that why we are using a pad_sequence

words_length = 8
embedding_doc =pad_sequences(one_hot_rep,padding='pre',maxlen=words_length)
print(embedding_doc)
```

```python
## feature representation
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=words_length))
model.compile('adam','mse')
```

```python
model.build(input_shape=(None, words_length))
```

```python
model.summary()
```

```python
model.predict(embedding_doc)
```
