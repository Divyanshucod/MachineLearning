```python
import pandas as pd

messages = pd.read_csv('spam1.csv',encoding='latin-1')
```

```python
messages = messages.iloc[:,0:2]
```

```python
messages.rename(columns={'v1': 'label', 'v2': 'message'}, inplace=True)
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
wordLemmatizer = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    # print(f'{review} : before split')
    review = review.split()
    # print(f'{review}: after split')
    review = [wordLemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review= ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=100)
```

```python
x= tfidf.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x:"%.3g" % x))
```
