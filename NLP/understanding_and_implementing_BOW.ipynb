```python
import pandas as pd
message = pd.read_csv('spam1.csv',encoding='latin-1')
```

```python
message = message.iloc[:,0:2]
```

```python
message.rename(columns={'v1': 'label', 'v2': 'message'}, inplace=True)
```

```python
message
```

```python
## data cleaning and preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(message)):
    review = re.sub('[^a-zA-Z]',' ',message['message'][i])
    review = review.lower()
    print(f'{review} : before split')
    review = review.split()
    print(f'{review}: after split')
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review= ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## creating bag model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500) 
## for binary bag of words : use binary:True in the CountVectorizer
# cv = CountVectorizer(binary=True)
```

```python
x = cv.fit_transform(corpus).toarray()
```
